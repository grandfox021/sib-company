```python
df = pd.read_excel("sales.xlsx") # To-Do
```



<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>    
    حذف فاکتور هایی که شناسه مشتری مشخصی ندارند
    
</font>
</p>


```python
df_no_missing_CustomerId = df.copy()


df_no_missing_CustomerId.drop(df_no_missing_CustomerId[df_no_missing_CustomerId['CustomerId'].isna()].index,inplace=True) # To-Do
df_no_missing_CustomerId
```

```python
df_no_duplicate = df_no_missing_CustomerId.copy()

df_no_duplicate = df_no_duplicate.drop_duplicates() # To-Do

df_no_duplicate
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
قسمت سوم    
</font>
</h2>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>        
سطر هایی که قیمت منفی دارن را حذف میکنیم
</font>
</p>



```python
df_no_missing_price = df_no_duplicate.copy()


df_no_missing_price = df_no_missing_price [df_no_missing_price.UnitPrice  > 0]# To-Do

df_no_missing_price
```


<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
   محاسبه اینکه چند درصد از کل سفارش ها لغو شده است
</font>



</font>
</p>

```python
percentage_cancelled_orders = df_no_missing_price.copy()
percentage_cancelled_orders = percentage_cancelled_orders.astype({'InvoiceNumber':'string'})
percentage_cancelled_orders = (percentage_cancelled_orders[percentage_cancelled_orders['InvoiceNumber'].str.contains('C')].nunique()[0]/ percentage_cancelled_orders['InvoiceNumber'].nunique()) *100
percentage_cancelled_orders = percentage_cancelled_orders.round() # To-Do
percentage_cancelled_orders
```



<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>  
پنج مشتری با بیشترین تعداد لغو سفارش
</font>
</p>


</p>

```python
top_cancelling_customers = df_no_missing_price.copy()
top_cancelling_customers = top_cancelling_customers.astype({'InvoiceNumber':'string'})
top_cancelling_customers = top_cancelling_customers.astype({'CustomerId':'int'}).astype({'CustomerId':'str'})
top_cancelling_customers = top_cancelling_customers[top_cancelling_customers['InvoiceNumber'].str.contains("C")].groupby('CustomerId') # To-Do
# top_cancelling_customers.size().sort_values(ascending=False)
top_cancelling_customers = top_cancelling_customers.nunique().sort_values(by='InvoiceNumber',ascending=False).head()
top_cancelling_customers = list(top_cancelling_customers.index)
top_cancelling_customers
```



<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>  
ردیف های مربوط به سفارش های لغو شده را حذف میکنیم
</font>
</p>


```python

df_no_canceled_invoice = df_no_missing_price.copy()
df_no_canceled_invoice = df_no_canceled_invoice.astype({'InvoiceNumber':'string'})
df_no_canceled_invoice.drop(df_no_canceled_invoice[df_no_canceled_invoice['InvoiceNumber'].str.contains('C')].index,inplace = True) # To-Do

df_no_canceled_invoice
```

```python
df_no_canceled_invoice.to_csv("preprocessed_sales.csv", index=False)
```

```python
remaining_invoices = pd.Series(df_no_canceled_invoice['InvoiceNumber'].unique())
remaining_invoices # To-Do
```
